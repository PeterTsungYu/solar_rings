### 爬取資料並將所有資料轉成 float 型態
### 將 WD 以 sin,cos 方式轉換，並將連續遺失6個值以內的特徵以差分方式補值

```python
# 將所有資料轉成 float 型態
def f(x):
    try:
        return np.float(x)
    except:
        return np.nan
```

```python
# 將 WD 以 sin,cos 方式轉換
def convert_to_sin(x):
    return np.sin((x/180)*np.pi)
def convert_to_cos(x):
    return np.cos((x/180)*np.pi)
```

```python
def codis(station, stname, date):
    params = {'command':'viewMain', 
              'station':station, 
              'stname':quote(stname), 
              'datepicker':date}
    resp = requests.get(url, params=params)
    resp.encoding = 'utf-8'
    soup = BeautifulSoup(resp.text, 'html.parser')
    MyTable = soup.find('table', id="MyTable")
    feat_len = len(MyTable.find('tr', class_="second_tr").select('th'))
    MyTable_values = MyTable.find_all('td')
    MyTable_len = int(len(MyTable.find_all('td'))/feat_len)
    lst_rows = [MyTable_values[(i*feat_len):((i+1)*feat_len)]for i in range(MyTable_len)]

    assert len(lst_rows) == MyTable_len

    for k in lst_rows:
        assert len(k) == feat_len

    StnPres = []
    Temperature = []
    RH = []
    WS = []
    WD = []
    Precp = []
    SunShine = []
    GloblRad = []

    for l in range(MyTable_len):
        StnPres.append(lst_rows[l][1].get_text(strip=True))
        Temperature.append(lst_rows[l][3].get_text(strip=True))
        RH.append(lst_rows[l][5].get_text(strip=True))
        WS.append(lst_rows[l][6].get_text(strip=True))
        WD.append(lst_rows[l][7].get_text(strip=True))
        Precp.append(lst_rows[l][10].get_text(strip=True))
        SunShine.append(lst_rows[l][12].get_text(strip=True))
        GloblRad.append(lst_rows[l][13].get_text(strip=True))

    df = pd.DataFrame(data={
                            'StnPres':StnPres, 
                            'Temperature':Temperature, 
                            'RH':RH, 
                            'WS':WS, 
                            'WD':WD, 
                            'Precp':Precp, 
                            'SunShine':SunShine, 
                            'GloblRad':GloblRad
                            }, 
                     dtype='float',
                     )
    
    # 'SunShine'欄位的 '...' 代表晚上所以轉成 0.0 
    for i in range(MyTable_len):
        if df.loc[i, 'SunShine'] == '...':
            df.loc[i, 'SunShine'] = 0.0
    
    # 把資料都轉成float，無法轉成float的就轉成NaN後面補值
    for j in df.columns:
        df[j] = df[j].apply(f)
        
    # WD是 0~360 的數值，直接補值會有錯誤的影響，於是轉成sin跟cos
    df['WD_sin'] = df['WD'].apply(convert_to_sin)
    df['WD_cos'] = df['WD'].apply(convert_to_cos)
    
    # 進行差分補值
    df = df.reindex(columns=['StnPres', 'Temperature', 'RH', 'WS', 'WD_sin', 'WD_cos', 'Precp', 'SunShine', 'GloblRad'])
    for m in df.columns:
        mask = df[m].isna()
        d = df.index.to_series()[mask].groupby((~mask).cumsum()[mask]).agg(['first', 'size'])
        d = d.rename(columns=dict(size='contig_null', first='Start_Date')).reset_index(drop=True)
        time_list = d[d.contig_null>6].Start_Date.to_list()
        conti_num = d[d.contig_null>6].contig_null.to_list()
        y = df[df.index.to_series().isin(time_list)].index
        
        df[m] = df[m].interpolate()
        
        indx_mask = []
        if len(conti_num)>0:
            for idx, k in enumerate(conti_num):
                indx_mask+=list(range(y[idx], y[idx]+k))
            df.loc[indx_mask, m] = np.nan
    
    return df
```

```python
# scall後的資料與各站點以時間合併
def StaCrawler(s_date, e_date):
    
    _24H = ['H0'+str(i) for i in range(1, 10)]+['H'+str(i) for i in range(10, 25)]
    sta_info = sta_dict.copy().iloc[:, 1:]
    #print(_24H)

    while s_date != e_date:
        
        for v in _24H:
            exec(f"{v} = pd.DataFrame(index=range(len(codes)), columns=['StnPres','Temperature','RH','WS', 'WD_sin', 'WD_cos', 'Precp','SunShine','GloblRad'])")
            #print(H01)
        
        #for i in range(3):
        for i in range(len(codes)):
            df = codis(codes[i], stations[i], s_date)
            #print(stations[i])

            for u in range(24):
                if u < 9:
                    locals()[f'H0{u+1}'].iloc[i] = df.iloc[u]
                else:
                    locals()[f'H{u+1}'].iloc[i] = df.iloc[u]
        
        for q in range(24):
            if q < 9:
                time_df = pd.concat((sta_info, locals()[f'H0{q+1}']), axis=1)
                time_df.to_csv(f'./CODiS/{s_date}_{_24H[q]}.csv', index=False)
            elif 9 <= q < 23:
                time_df = pd.concat((sta_info, locals()[f'H{q+1}']), axis=1)
                time_df.to_csv(f'./CODiS/{s_date}_{_24H[q]}.csv', index=False)
            else:
                s_date = str((pd.to_datetime(s_date)+pd.Timedelta(1, unit='d')).date())
                time_df = pd.concat((sta_info, locals()[f'H{q+1}']), axis=1)
                time_df.to_csv(f'./CODiS/{s_date}_H00.csv', index=False)
```

```python
# merge all station by time

_24H = ['H0'+str(i) for i in range(1, 10)]+['H'+str(i) for i in range(10, 25)]
s_date = '2017-01-01'
e_date = '2017-02-01'

df_tmp = pd.DataFrame()
while s_date != e_date:
    for i in range(24):
        if i < 9:
            csv_name = f'./CODiS/{s_date}_{_24H[i]}.csv'
        elif 9 <= i < 23:
            csv_name = f'./CODiS/{s_date}_{_24H[i]}.csv'
        else:
            s_date = str((pd.to_datetime(s_date)+pd.Timedelta(1, unit='d')).date())
            csv_name = f'./CODiS/{s_date}_H00.csv'
            
        df = pd.read_csv(csv_name, usecols=['StnPres', 'Temperature', 'RH', 'WS', 'WD_sin', 'WD_cos', 'Precp', 'SunShine', 'GloblRad'])
        df_tmp = pd.concat((df_tmp, df))
```

### 各個時間的資料以 KNN 的方式補值

```python
time_path = sorted(glob.glob('CODiS/*.csv'))

for j in tqdm.tqdm_notebook(time_path):
    
    print(j, end='\r', flush=True)
    df = pd.read_csv(j)
    scaler = MinMaxScaler(feature_range=(0,1))
        
    # 7:14 ['StnPres', 'Temperature', 'RH', 'WS', 'WD_sin', 'WD_cos', 'Precp']
    df.iloc[:, 7:14] = scaler.fit_transform(df.iloc[:,7:14])
    imputer = KNNImputer(n_neighbors=3)
      
    # 4:14 ['altitude_s', 'latitude_s', 'longitude_s', 'StnPres', 'Temperature', 'RH', 'WS', 'WD_sin', 'WD_cos', 'Precp']
    df.iloc[:, 4:14] = pd.DataFrame(imputer.fit_transform(df.iloc[:, 4:14]), columns=df.columns[4:14])
    df.to_csv(f'fill_{j}', index=False)
```

```python
time_path = sorted(glob.glob('fill_CODiS/fill_Station_bytime_2017/*.csv'))

for j in tqdm.tqdm_notebook(time_path):
    
    print(j, end='\r', flush=True)
    df = pd.read_csv(j)[:9]
    scaler = MinMaxScaler(feature_range=(0,1))
        
    # 7:14 ['StnPres', 'Temperature', 'RH', 'WS', 'WD_sin', 'WD_cos', 'Precp']
    df.iloc[:, 14:16] = scaler.fit_transform(df.iloc[:,14:16])
    imputer = KNNImputer(n_neighbors=3)
    
    df.iloc[:, 14:16] = scaler.inverse_transform(df.iloc[:, 14:16])
      
    # 4:14 ['altitude_s', 'latitude_s', 'longitude_s', 'StnPres', 'Temperature', 'RH', 'WS', 'WD_sin', 'WD_cos', 'Precp']
    df.iloc[:, 4:16] = pd.DataFrame(imputer.fit_transform(df.iloc[:, 4:16]), columns=df.columns[4:16])
    df.to_csv(f'fill_CODiS/fill2_Station_bytime_2017/{j[-18:]}', index=False)
```
